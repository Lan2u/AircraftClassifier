```python
import os.path
import psycopg2

import configparser
```

```python
IMAGE_VARIANT_FILE = "aircraft-data/data/images_variant_trainval.txt"
IMAGES_FOLDER = "aircraft-data/data/images"
```

```python
class DBConfig:
    def __init__(self, filename, host, port, db_name, user, pwd):
        self.filename = filename
        self.host = host
        self.port = port
        self.db_name = db_name
        self.user = user
        self.pwd = pwd
```

```python
def load_config(filename = 'database.ini', section = 'aircraft_postgres_db'):
    parser = configparser.ConfigParser()
    parser.read(filename)
    cfg = parser[section]
    return DBConfig(filename, cfg['Host'], cfg['Port'], cfg['Db_name'], cfg['User'], cfg['Pass'])
```

```python
load_config().pwd
```

```python
conn = psycopg2.connect()
```

```python
cur = conn.cursor()
cur.execute('SELECT version()')
print("Connected to database, version: " + str(cur.fetchone()))
```

```python
with open(IMAGE_VARIANT_FILE, "r") as variant_file:
    for variant_line in variant_file:
        parts = variant_line.split(" ", 1)
        image_name = parts[0]
        variant = parts[1]
        image_path = os.path.join(IMAGES_FOLDER, image_name + '.jpg')
        store_image_details(image_name, image_path, variant)
        
```
